# Stemming words

```python
from nltk.stem import PorterStemmer  
ps = PorterStemmer()   
 
def Stemming_words(msg):
    stem_lis = []
    for words in msg:
        lis_words = words.split()  
        stem_words = ""
        count_word = 0
        for word in lis_words:
            stem_str = ps.stem(word)  
            stem_words += stem_str  
            count_word += 1
            if len(lis_words) != count_word:
                stem_words += " "
        stem_lis.append(stem_words)
    return stem_lis
stem_lis = Stemming_words(corpus)
```

# Removing Punctuations

```python
from string import punctuation as punc
def Removing_punctuations(msg):
    rev = []
    for rw in msg:     
        no_punct = ""    
        for char in rw:
            if char not in punc:
                no_punct = no_punct + char
        rev.append(no_punct)
    return rev
msg = Removing_punctuations(stem_lis)  
```

# Removing Stopwords

```python
from sklearn.feature_extraction import stop_words    
from nltk.tokenize import word_tokenize
stop_words = list(stop_words.ENGLISH_STOP_WORDS)
      
def Removinig_stopwords(msg):  
    sentences = []
    for up in msg:
        word_tokens = word_tokenize(up)
        filtered_sentence = []
        for tokens in word_tokens:
            if tokens not in stop_words:
                filtered_sentence.append(tokens)         
        str_sentence = ""
        for s in filtered_sentence:
            str_sentence += s
            str_sentence += " "
        sentences.append(str_sentence)
    return sentences   

massages = Removinig_stopwords(msg)
```

# vectorizing with tfidf vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer      
vec = TfidfVectorizer()
matrix_X = vec.fit_transform(massages)     
```

```python
from sklearn.model_selection import train_test_split     
traning_doc,testing_doc,traning_lab,testing_lab = train_test_split(matrix_ X ,lables ,train_size=0.9 ,shuffle = True)
```

# Traning MultinomialNB classifier

```python
from sklearn.naive_bayes import MultinomialNB   
clf = MultinomialNB()
clf.fit(traning_doc,traning_lab)
pred_labels = clf.predict(testing_doc)   
```
